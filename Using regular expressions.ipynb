```python
import re
file_name="pride.txt"  
with open(file_name,encoding='utf-8') as f: 
    words=f.read()
print(words)
count5=re.findall(r'\b[a-zA-Z]{5}\b', words)  
count10=re.findall(r'\w[a-zA-Z]{10}\w', words)  
capitalized=re.findall('([A-Z][a-z]+)', words)   
lower_case=re.findall(r"[a-z]+", words)   
print(count5,count10,capitalized,lower_case)
print("Count of:")
print("5 words=",len(count5))  
print("10 words=",len(count10))
print("Capitalized words=",len(capitalized))
print("Lower case words=",len(lower_case))
```
